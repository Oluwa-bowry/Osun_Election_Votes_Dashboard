```python
import pandas as pd
import numpy as np
```

```python
 #Assuming your dataset is loaded into a DataFrame called 'df'
 # If not, you can load it with: 
gdf_sorted = pd.read_csv('OSUN_geocoded_cleaned_final.csv')

 # Step 1: Calculate z-scores for each party's votes
 # We’ll calculate the z-score for APC, LP, PDP, and NNPP votes
 parties = ['APC', 'LP', 'PDP', 'NNPP']
 for party in parties:
    # Calculate mean and standard deviation for the party's votes
    mean = gdf_sorted[party].mean()
    std = gdf_sorted[party].std()
    
    # Calculate z-score for each polling unit's votes for this party
    # Handle cases where std is 0 to avoid division by zero
    if std != 0:
        gdf_sorted[f'{party}_z_score'] = (gdf_sorted[party] - mean) / std
    else:
        gdf_sorted[f'{party}_z_score'] = 0  # If std is 0, set z-score to 0 (no 
# Step 2: Identify outliers based on z-scores
 # A common threshold for outliers is |z-score| > 3
 for party in parties:
     gdf_sorted[f'{party}_outlier'] = np.where(
         (gdf_sorted[f'{party}_z_score'].abs() > 3),  # Check if absolute z-score
         'Outlier',  # Label as outlier
         'Normal'    
        # Label as normal
    )
 # Step 3: Display the results
 # Show the polling units with their z-scores and outlier flags
print("Polling Units with Z-Scores and Outlier Flags for Each Party:")
display_columns = ['State', 'LGA', 'Ward', 'PU-Name', 'APC', 'APC_z_score', 'APC_outlier',
'LP', 'LP_z_score', 'LP_outlier', 'PDP', 'PDP_z_score', 'PDP_outlier',
'NNPP', 'NNPP_z_score', 'NNPP_outlier']
display(gdf_sorted[display_columns].head(10))  # Show first 10 rows for brevity
gdf_sorted.to_csv("OSUN_geocoded_cleaned_final.csv", index=False)
     
```
